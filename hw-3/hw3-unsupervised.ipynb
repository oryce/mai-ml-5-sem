**Задание 1.** (1 балл)

Сделайте кластеризацию данных методом `KMeans` (собственная имплементация и готовая реализация), обучив модель на данных со сниженной за счет PCA размерностью. В данном случае мы подскажем, что нужно искать именно 6 кластеров, но в общем случае мы не будем знать, сколько кластеров надо искать.

Параметры:

- **n_clusters** = n_classes (число уникальных меток целевого класса)
- **n_init** = 100
- **random_state** = RANDOM_STATE (для воспроизводимости результата)

Остальные параметры со значениями по умолчанию.

````python
class MyKMeans:
    def __init__(self, n_clusters=6, max_iter=300, random_state=17, n_init=10):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.random_state = random_state
        self.n_init = n_init

    def fit(self, X):
        np.random.seed(self.random_state)
        best_inertia = np.inf
        best_labels = None
        best_centers = None
        for init in range(self.n_init):
            centers = X[np.random.choice(X.shape[0], self.n_clusters, replace=False)]
            for iter in range(self.max_iter):
                distances = np.linalg.norm(X[:, np.newaxis] - centers, axis=2)
                labels = np.argmin(distances, axis=1)
                new_centers = np.array(
                    [
                        X[labels == k].mean(0) if np.any(labels == k) else centers[k]
                        for k in range(self.n_clusters)
                    ]
                )
                if np.allclose(centers, new_centers):
                    break
                centers = new_centers
            inertia = np.sum((X - centers[labels]) ** 2)
            if inertia < best_inertia:
                best_inertia = inertia
                best_labels = labels
                best_centers = centers
        self.labels_ = best_labels
        self.cluster_centers_ = best_centers
        self.inertia_ = best_inertia
        return self


# Собственная реализация
my_kmeans = MyKMeans(n_clusters=n_classes, random_state=RANDOM_STATE, n_init=100)
my_kmeans.fit(X_pca)
my_cluster_labels = my_kmeans.labels_

# Готовая реализация из sklearn
kmeans = KMeans(n_clusters=n_classes, n_init=100, random_state=RANDOM_STATE)
cluster_labels = kmeans.fit_predict(X_pca)

# Используем sklearn labels для дальнейших шагов
print("MyKMeans inertia:", my_kmeans.inertia_)
print("Sklearn KMeans inertia:", kmeans.inertia_)
````

````python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, s=20, cmap='viridis')
````

````python
tab = pd.crosstab(y, cluster_labels, margins=True)
tab.index = ['ходьба', 'подъем вверх по лестнице',
             'спуск по лестнице', 'сидение', 'стояние', 'лежание', 'все']
tab.columns = ['cluster' + str(i + 1) for i in range(6)] + ['все']
tab
````

````python
inertia = []
for k in tqdm_notebook(range(1, n_classes + 1)):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=RANDOM_STATE)
    kmeans.fit(X_pca)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, n_classes + 1), inertia)
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow method')
plt.show()
````
